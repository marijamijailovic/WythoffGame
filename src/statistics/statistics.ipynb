```python
import pandas as pd
import matplotlib.pyplot as plt 
```

```python
calculate_time_n = pd.read_csv('./csv/calculate_time_n.csv')
calculate_time_n.info()
```

```python
plt.figure(figsize=(1000/100,400/100), dpi=100);
plt.plot(calculate_time_n['n'][:18], calculate_time_n['recursive'][:18], color='blue')
plt.title('График рекурзивне стратегије за конструкцију П табеле')
plt.xlabel('n, a = 2')
plt.ylabel('Време извршавања(ms)')
plt.grid(True)
plt.annotate('max pair (1 853 640, 4 475 080)', xy=(1310720, 11513000), xytext=(1100000, 8000000),
             arrowprops=dict(facecolor='blue', shrink=0.05),
             )
plt.savefig('./picture/recursive.png', dpi=100)
plt.show()
```

```python
plt.figure(figsize=(1000/100,400/100), dpi=100);
plt.plot(calculate_time_n['n'], calculate_time_n['algebraic'], color='red')
plt.title('График алгебарске стратегије за конструкцију П табеле')
plt.xlabel('n, a = 2')
plt.ylabel('Време извршавања(ms)')
plt.grid(True)
plt.annotate('max pair (949 063 000, 2 291 240 000)', xy=(671088640, 42270.9), xytext=(538500000, 30000),
             arrowprops=dict(facecolor='red', shrink=0.05),
             )
plt.savefig('./picture/algebraic.png', dpi=100)
plt.show()
```

```python
plt.figure(figsize=(1000/100,400/100), dpi=100);
plt.plot(calculate_time_n['n'][:22], calculate_time_n['arithmetic'][:22], color='green')
plt.title('График аритметичке стратегије за конструкцију П табеле')
plt.xlabel('n, a = 2')
plt.ylabel('Време извршавања(ms)')
plt.grid(True)
plt.annotate('max pair (29 658 200, 71 601 200)', xy=(20971520, 230089.0), xytext=(17400000, 150000),
             arrowprops=dict(facecolor='green', shrink=0.05),
             )
plt.savefig('./picture/arithmetic.png', dpi=100)
plt.show()
```

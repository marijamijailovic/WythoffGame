```python
import pandas as pd
import matplotlib.pyplot as plt 
```

```python
calculate_time_n = pd.read_csv('./csv/calculate_time_n.csv')
calculate_time_n
```

```python
calculate_piles = pd.read_csv('./csv/calculate_piles.csv')
calculate_piles
```

```python
plt.figure(figsize=(1000/100,400/100), dpi=100);
plt.plot(calculate_time_n['n'], calculate_time_n['rekurzivna'], color='blue')
plt.title('График рекурзивне стратегије за конструкцију табеле изгзбљених позиција')
plt.xlabel('n, a = 2')
plt.ylabel('Време извршавања(ms)')
plt.grid(True)
plt.annotate('max pair (59 316 416, 143 202 496)', xy=(41885550, 10052.7), xytext=(34385550, 6500),
             arrowprops=dict(facecolor='blue', shrink=0.05),
             )
plt.savefig('./picture/recursive.png', dpi=100)
plt.show()
```

```python
plt.figure(figsize=(1000/100,400/100), dpi=100);
plt.plot(calculate_time_n['n'], calculate_time_n['algebarska'], color='red')
plt.title('График алгебарске стратегије за конструкцију табеле изгзбљених позиција')
plt.xlabel('n, a = 2')
plt.ylabel('Време извршавања(ms)')
plt.grid(True)
plt.annotate('max pair (59 316 416, 143 202 496)', xy=(41885550, 3065.81), xytext=(34385550, 2100),
             arrowprops=dict(facecolor='red', shrink=0.05),
             )
plt.savefig('./picture/algebraic.png', dpi=100)
plt.show()
```

```python
plt.figure(figsize=(1000/100,400/100), dpi=100);
plt.plot(calculate_time_n['n'], calculate_time_n['aritmeticka'], color='green')
plt.title('График аритметичке стратегије за конструкцију табеле изгзбљених позиција')
plt.xlabel('n, a = 2')
plt.ylabel('Време извршавања(ms)')
plt.grid(True)
plt.annotate('max pair (59 316 416, 143 202 496)', xy=(41885550, 587616), xytext=(34385550, 410000),
             arrowprops=dict(facecolor='green', shrink=0.05),
             )
plt.savefig('./picture/arithmetic.png', dpi=100)
plt.show()
```

```python
plt.figure(figsize=(1000/100,400/100), dpi=100);
plt.plot(calculate_time_n['n'],calculate_time_n['rekurzivna'],color='blue')
plt.plot(calculate_time_n['n'],calculate_time_n['algebarska'],color='red')
plt.plot(calculate_time_n['n'],calculate_time_n['aritmeticka'],color='green')
plt.legend()
plt.title('Сумиран приказ извршавања свих стратегија за конструкцију табеле изгзбљених позиција')
plt.xlabel('n, a = 2')
plt.ylabel('Време извршавања(ms)')
plt.grid(True)
plt.savefig('./picture/all.png', dpi=100)
plt.show()
```

```python
plt.figure(figsize=(1000/100,400/100), dpi=100);
plt.plot(calculate_time_n['n'],calculate_time_n['rekurzivna'],color='blue')
plt.plot(calculate_time_n['n'],calculate_time_n['algebarska'],color='red')
plt.legend()
plt.title('Сумиран приказ извршавања рекурзивне и алгебарске стратегије за конструкцију табеле изгзбљених позиција')
plt.xlabel('n, a = 2')
plt.ylabel('Време извршавања(ms)')
plt.grid(True)
plt.savefig('./picture/algebraicVSrecursive.png', dpi=100)
plt.show()
```
